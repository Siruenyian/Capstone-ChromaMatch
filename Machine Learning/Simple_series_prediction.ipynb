```python
import numpy as np
import tensorflow as tf
import pandas as pd
```

```python
# Define your input sequence
input_sequence = np.array([[1, 2, 3]])

# Define your target sequence
target_sequence = np.array([[10, 11, 12], [13, 14, 15], [16, 17, 18]])


```

```python
input_sequence = np.reshape(input_sequence, (input_sequence.shape[0], input_sequence.shape[1], 1))
target_sequence = np.reshape(target_sequence, (1,-1))
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.LSTM(50, activation='relu', input_shape=(input_sequence.shape[1], input_sequence.shape[2])))
model.add(tf.keras.layers.Dense(target_sequence.shape[1]))

```

```python
# Compile the model
model.compile(optimizer='adam', loss='mse')

```

```python
model.summary()
```

```python
# Train the model
model.fit(input_sequence, target_sequence, epochs=100, batch_size=1)

```

```python
# Generate some test data
test_input = np.array([[10, 12, 13]])


predictions = model.predict(test_input)

# Print the predictions
print(predictions)
```
